```python
import matplotlib.pyplot as plt
```

```python
year = [2012, 2013, 2014, 2015, 2016, 2017]
salary = [12, 13, 14, 17, 19, 20]
population = [100, 120, 180, 250, 300, 370]
```

### plt.scatter(x_axis, y_axis, size/s, color/c)

```python
# plt.scatter(year, salary)
plt.scatter(year, salary, s = population, c = 'r')
plt.title("Bubble Chart")
plt.show()
```

```python
plt.scatter(year, salary, s = population, c = 'r', alpha=0.5)
plt.title("Bubble Chart")
plt.show()
```

```python
import numpy as np
c = np.arange(len(year))
```

```python
plt.scatter(year, salary, s = population, c = c, alpha=0.5)
plt.title("Bubble Chart")
plt.show()
```

```python
### edge color
plt.scatter(year, salary, s = population, c = c, alpha=0.5, edgecolors='black')
plt.title("Bubble Chart")
plt.show()
```

```python
### chaange marker
plt.scatter(year, salary, s = population, c = c, alpha=0.5, edgecolors='black', marker='^')
plt.title("Bubble Chart")
plt.show()
```

```python
### edge color
plt.scatter(year, salary, s = population, c = c, alpha=0.5, edgecolors='black')
plt.title("Bubble Chart")
plt.text(year[0], salary[0], "100")
plt.show()
```

```python
### edge color
plt.scatter(year, salary, s = population, c = c, alpha=0.5, edgecolors='black')
plt.title("Bubble Chart")
plt.text(year[0] + 0.1, salary[0] + 0.1, population[0])
plt.text(year[4] + 0.1, salary[4] + 0.1, population[4])
plt.show()
```

```python
### edge color
plt.scatter(year, salary, s = population, c = c, alpha=0.5, edgecolors='black')
plt.title("Bubble Chart")
for i in range(len(year)):
    plt.text(year[i] + 0.1, salary[i] + 0.1, population[i])
plt.show()
```
