### Color, Marker, linewidth, labels for axis, datasets, title, axis, grid

### plt.axis([x_start, x_end, y_start, y_end])

### plt.grid()

### plt.text(x, y, "text")

```python
import numpy as np
# x = np.array([1,2,3,4])
x = np.arange(0, 5, 0.2)
y = x**3
y2 = x**2

plt.axis([0, 10, 0, 100])

# plt.plot(x, y, color = "black", marker = "+")
plt.plot(x, y, color = "black", linewidth = 2, label = "x^3")
plt.plot(x, y2, color = "red", linewidth = 2, label = "x^2")
plt.ylabel("X^3")
plt.xlabel("X")
plt.legend()
plt.title("MatplotLib demo")
plt.text(2, 80, "text", fontsize = 12)
plt.grid()
plt.show()
```

```python
x = np.array([i*2 for i in range(20)])
y = x

# plt.plot(x, color = 'red', label = "X")
plt.plot(y, color = 'green', label = "Y")
plt.plot(x, color = 'red', label = "X")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

plt.show()
```
