```python
#Print Total wounded people
import csv
with open('year2017.csv', 'r') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)
    print(type(file_data))
    # file_list = list(file_data)
    
    wound = []
    
    for row in file_data:
        val = row['Wounded']
        if val != '':
            wound.append(int(float(val)))
    
    print(int(sum(wound)))
```

```python
#Print Total wounded people from India
import csv
with open('year2017.csv', 'r') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)
    print(type(file_data))
    file_list = list(file_data)
    
    wound = {}
    
    for row in file_list:
        key = row['Country']
        val = row['Wounded']
        
        if val != '':
            val = int(float(val))
        else:
            val = 0
        
        if key in wound:
            wound[key] += val
        else:
            wound[key] = val
    
    print(wound['India'])
```

```python
# Casualties from Explosives type
import csv
with open('year2017.csv', 'r') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)
    file_list = list(file_data)
    # casualties = {}
    sum = 0
    for row in file_list:
        if row['Weapon_type'] == 'Explosives' and row['casualities'] != '':
            sum += int(float(row['casualities']))
            
    print(sum)
```

```python
# Total number of people killed per month
import csv
with open('year2017.csv', 'r') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)
    file_list = list(file_data)
    KilledPerMonth = {}
    for row in file_list:
        key = row['Month']
        val = row['Killed']
        if row['Killed'] != '':
            val = int(float(val))
        else:
            val = 0
            
        if key in KilledPerMonth:
            KilledPerMonth[key] += val
        else:
            KilledPerMonth[key] = val
            
    # print(KilledPerMonth)
KilledPerMonth
```

```python
import csv
with open('year2017.csv', 'r') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)
    file_list = list(file_data)
    
    CasualtiesPerCountry = {}
    for row in file_list:
        key = row['Country']
        val = row['casualities']
        
        if val != '':
            val = int(float(val))
        else:
            val = 0
            
        if key in CasualtiesPerCountry:
            CasualtiesPerCountry[key] += val
        else:
            CasualtiesPerCountry[key] = val
            
    print(CasualtiesPerCountry)
```
