```python
import csv
with open('amazon_jobs_dataset.csv', 'r') as file_obj:
    file_data = csv.DictReader(file_obj)
    file_list = list(file_data)
    
    job_opening = {}
    for row in file_list:
        key = row['location']
        if key in job_opening:
            job_opening[key] = job_opening[key] + 1
        else:
            job_opening[key] = 1
            
    # print(job_opening)
    print('US, WA, Seattle ', job_opening['US, WA, Seattle '])
    print('IN, KA, Bangalore', job_opening['IN, KA, Bangalore '])
```

```python
import csv
with open('amazon_jobs_dataset.csv', 'r') as file_obj:
    file_data = csv.DictReader(file_obj)
    file_list = list(file_data)
    
    jobs = {}
    
    for row in file_list:
        key = row['Posting_date']
        # key.__contains__('Computer Vision')
        if key.endswith('2018'):
            list_str = key.split()
            if list_str[0] in jobs:
                jobs[list_str[0]] = jobs[list_str[0]] + 1
            else:
                jobs[list_str[0]] = 1
                
    # print(jobs)  
    s = ''
    mx = 0
    for d in jobs:
        # print(jobs[d])
        if jobs[d] > mx:
            mx = jobs[d]
            s = d
    
    print(s, mx)
            
```

```python
import csv
with open('amazon_jobs_dataset.csv', 'r') as file_obj:
    file_data = csv.DictReader(file_obj)
    file_list = list(file_data)
    
    jobs_for_bachelor = 0
    
    for row in file_list:
        key = row['BASIC QUALIFICATIONS']
        # key.__contains__('Computer Vision')
        if key.__contains__('Bachelor') or key.__contains__('BS') or key.__contains__('BA'):
            # print(key)
            jobs_for_bachelor = jobs_for_bachelor + 1
        
    print(jobs_for_bachelor)
```

```python
import csv
with open('amazon_jobs_dataset.csv', 'r') as file_obj:
    file_data = csv.DictReader(file_obj)
    file_list = list(file_data)
    
    jobs_based_language = 0
    
    for row in file_list:
        key = row['BASIC QUALIFICATIONS']
        loc = row['location']
        # key.__contains__('Computer Vision')
        if loc.startswith('IN'):
            if key.__contains__('Bachelor') or key.__contains__('BS') or key.__contains__('BA'):
                if 'Java' in key or 'C++' in key or 'Python' in key:
                    jobs_based_language = jobs_based_language + 1
        
    print(jobs_based_language)
```
